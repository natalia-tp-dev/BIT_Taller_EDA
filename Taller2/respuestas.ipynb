```python
def tipificarFechas(df, columnas):
    for columna in columnas:
        df[columna] = pd.to_datetime(df[columna], errors='coerce')

def tipificarCodigo(df, columnas):
    for columna in columnas:
        df[columna] = df[columna].astype("object")

tipificarCodigo(df, ["Postal Code"])
tipificarFechas(df, ["Order Date", "Ship Date"])
df.dtypes
```

## **Moda / Rango / Desviación estándar / Varianza**

```python
# defino mis columnas numéricas para no mezclar otras
num_cols = ["Sales", "Quantity", "Ship Cost", "Taxes"]
num_vars = df[num_cols]

# moda: valor que más se repite
# uso .iloc[0] porque .mode() puede traer más de una moda; me quedo con la primera para dejar 1 valor por columna
moda = num_vars.mode().iloc[0]

# rango: max - min (medida rápida de dispersión, me da idea del “ancho” de los datos)
rango = num_vars.max() - num_vars.min()

# desviación estándar: qué tanto se mueven los datos alrededor del promedio
desv = num_vars.std()

# varianza: dispersión al cuadrado (menos intuitiva, pero útil para comparar dispersiones)
var_ = num_vars.var()

# armo una tabla para verlo ordenado; .round(3) es solo para que se vea limpio (3 decimales)
resumen = pd.DataFrame({
    "Moda": moda,
    "Rango": rango,
    "Desviación estándar": desv,
    "Varianza": var_
}).round(3)

print("Resumen de estadísticas univariadas:")
print(resumen)
```

## **Tablas de contingencia / dinámicas**

```python
# tabla de conteos -> cuántos registros hay de cada combinación
tabla_segment_shipmode = pd.crosstab(df["Segment"], df["Ship Mode"])
print("Conteos Segment vs Ship Mode:")
print(tabla_segment_shipmode)

# tabla de porcentajes por fila -> distribución de Ship Mode dentro de cada Segment
tabla_segment_shipmode_pct = pd.crosstab(df["Segment"], df["Ship Mode"], normalize="index").round(3)
print("Porcentajes por fila Segment vs Ship Mode:")
print(tabla_segment_shipmode_pct)
```

### Por qué esta tabla?: 

permite ver qué tipos de clientes prefieren ciertos modos de envío.

Ejemplo: si los clientes Corporate usan más “Second Class” que los Consumer.

Los porcentajes ayudan a interpretar mejor que los conteos absolutos.

```python
# tabla dinámica -> resumen de ventas y cantidades por categoría
pivot_sales_quantity_category = pd.pivot_table(
    df,
    index="Category",
    values=["Sales", "Quantity"],
    aggfunc=["mean", "sum", "std"]  # promedio, suma y dispersión
).round(2)

print("Pivot de Sales y Quantity por Category:")
print(pivot_sales_quantity_category)
```

### Por qué esta tabla:

Resume cómo se comporta cada categoría de producto en términos de dinero (Sales) y volumen (Quantity).

Permite ver cuál categoría aporta más ventas totales y cuál mueve más unidades.

Es una forma clara de relacionar una categórica (Category) con numéricas (Sales, Quantity).
