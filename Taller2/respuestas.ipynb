- Tipificamos las fechas como datetime y el codigo postal lo volvemos un objeto tipo string

```python
def tipificarFechas(df, columnas):
    for columna in columnas:
        df[columna] = pd.to_datetime(df[columna], errors='coerce')

def tipificarCodigo(df, columnas):
    for columna in columnas:
        df[columna] = df[columna].astype("object")

tipificarCodigo(df, ["Postal Code"])
tipificarFechas(df, ["Order Date", "Ship Date"])
df.dtypes
```

## **Moda / Rango / Desviaci√≥n est√°ndar / Varianza**

```python
# defino mis columnas num√©ricas para no mezclar otras
num_cols = ["Sales", "Quantity", "Ship Cost", "Taxes"]
num_vars = df[num_cols]

# moda: valor que m√°s se repite
# uso .iloc[0] porque .mode() puede traer m√°s de una moda; me quedo con la primera para dejar 1 valor por columna
moda = num_vars.mode().iloc[0]

# rango: max - min (medida r√°pida de dispersi√≥n, me da idea del ‚Äúancho‚Äù de los datos)
rango = num_vars.max() - num_vars.min()

# desviaci√≥n est√°ndar: qu√© tanto se mueven los datos alrededor del promedio
desv = num_vars.std()

# varianza: dispersi√≥n al cuadrado (menos intuitiva, pero √∫til para comparar dispersiones)
var_ = num_vars.var()

# armo una tabla para verlo ordenado; .round(3) es solo para que se vea limpio (3 decimales)
resumen = pd.DataFrame({
    "Moda": moda,
    "Rango": rango,
    "Desviaci√≥n est√°ndar": desv,
    "Varianza": var_
}).round(3)

print("Resumen de estad√≠sticas univariadas:")
print(resumen)
```

## **Tablas de contingencia / din√°micas**

```python
# tabla de conteos -> cu√°ntos registros hay de cada combinaci√≥n
tabla_segment_shipmode = pd.crosstab(df["Segment"], df["Ship Mode"])
print("Conteos Segment vs Ship Mode:")
print(tabla_segment_shipmode)

# tabla de porcentajes por fila -> distribuci√≥n de Ship Mode dentro de cada Segment
tabla_segment_shipmode_pct = pd.crosstab(df["Segment"], df["Ship Mode"], normalize="index").round(3)
print("Porcentajes por fila Segment vs Ship Mode:")
print(tabla_segment_shipmode_pct)
```

### Por qu√© esta tabla?: 

permite ver qu√© tipos de clientes prefieren ciertos modos de env√≠o.

Ejemplo: si los clientes Corporate usan m√°s ‚ÄúSecond Class‚Äù que los Consumer.

Los porcentajes ayudan a interpretar mejor que los conteos absolutos.

```python
# tabla din√°mica -> resumen de ventas y cantidades por categor√≠a
pivot_sales_quantity_category = pd.pivot_table(
    df,
    index="Category",
    values=["Sales", "Quantity"],
    aggfunc=["mean", "sum", "std"]  # promedio, suma y dispersi√≥n
).round(2)

print("Pivot de Sales y Quantity por Category:")
print(pivot_sales_quantity_category)
```

### Por qu√© esta tabla:

Resume c√≥mo se comporta cada categor√≠a de producto en t√©rminos de dinero (Sales) y volumen (Quantity).

Permite ver cu√°l categor√≠a aporta m√°s ventas totales y cu√°l mueve m√°s unidades.

Es una forma clara de relacionar una categ√≥rica (Category) con num√©ricas (Sales, Quantity).
